## Pandas

Pandas is a powerful data manipulation library in python, widely used for data analysis and data cleaning. It provides two primary data structures: series and dataframe. A Series is one_dimensional array-like object, while DataFrame is two dimensional, size-mutable and potentially heterogeneous tabular data structure with labeled axes(rows and columns)

```python
import pandas as pd
```

```python
##Series: one dimensional array-like object that can hold any data type. It is similar to column in a table.

import pandas as pd
data=[1,2,3,4,5]
series=pd.Series(data)
print("Series \n", series)
print(type(series))
```

```python
## Series from dictionary elements
data={'a':1,'b':2,'c':3}
series_dict = pd.Series(data)
print(series_dict)
```

```python
data = [10,20,30]
index=['a','b','c']
pd.Series(data,index=index)
```

```python
## DataFrame
## Create a dataframe from a dictionary of list

data = {
    'Name' : ['Krish','Prerna','Mayank'],
    'Age' : [25,30,45],
    'City' : ['Banglore','Noida', 'Gurgaon']
}
df=pd.DataFrame(data)
print(df)
print(type(df))
```

```python
import numpy as np
np.array(df)
```

```python
## Create a data frame fron list of dictionaries
data = [
    {'Name' : 'Krish','Age':32, 'City':'Banglore'},
    {'Name' : 'Kriti','Age':35, 'City':'Noida'},
    {'Name' : 'Krishna','Age':42, 'City':'Mumbai'},
    {'Name' : 'Kripa','Age':52, 'City':'Lucknow'}
]
df=pd.DataFrame(data)
print(df)
print(type(df))
```
